```python
train_setimen = data['train_text'].values
label_sentimen = data[['anger','fear','joy','sadness']].values
```

```python
import numpy as np

# Contoh array label multilabel
label_multilabel = np.array([[1, 0, 0, 0],  # Contoh untuk label "anger"
                             [0, 1, 0, 0],  # Contoh untuk label "fear"
                             [0, 0, 1, 0],  # Contoh untuk label "joy"
                             [0, 0, 0, 1]]) # Contoh untuk label "sadness"

# Menggunakan np.argmax untuk menemukan kelas yang sesuai
kelas_prediksi = np.argmax(label_multilabel, axis=1)

print(kelas_prediksi)
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test, label_train, label_test = train_test_split(train_setimen, label_sentimen, test_size=0.2)
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
```

```python
tokenizer = Tokenizer(5000, oov_token='x')
tokenizer.fit_on_texts(data_train)
tokenizer.fit_on_texts(data_test)
```

```python
sekuen_latih = tokenizer.texts_to_sequences(data_train)
padded_latih = pad_sequences(sekuen_latih)
```

```python

sekuen_test = tokenizer.texts_to_sequences(data_test)
padded_test = pad_sequences(sekuen_test)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(5000, output_dim=16),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(126, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax'),
])

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(padded_latih, label_train, epochs=30, verbose=2, validation_data=(padded_test, label_test))
```

```python
# meuni asa rame nya crime dimamana... kemarin juga rampok di Domino Pizza Margahayu, siang-siang deuih eta mah bawa senjata api. Sarieun kieu nya Bandung
def prediksi(test):
  teks_seq = tokenizer.texts_to_sequences(test)
  teks_padded = pad_sequences(teks_seq)
  return teks_padded

predict= model.predict(prediksi(["aduuuhhh... hate kuring meuni bagja pisan jeung adem ngadangukeun tembang na"]))

np.argmax(predict)
```

```python
def prediksi(test):
    teks_seq = tokenizer.texts_to_sequences(test)
    teks_padded = pad_sequences(teks_seq, padding='post')
    return teks_padded

# Membaca data dari file CSV
data = pd.read_csv('test.csv')  # Ganti dengan path ke file CSV Anda

# Mengambil kolom teks untuk diprediksi
texts = data['data'].values  # Ganti 'text' dengan nama kolom yang sesuai di file CSV Anda

# Memproses teks dan melakukan prediksi
padded_texts = prediksi(texts)
predictions = model.predict(padded_texts)

# Mengambil label dengan probabilitas tertinggi
predicted_labels = np.argmax(predictions, axis=1)

# Menambahkan hasil prediksi ke DataFrame
data['predicted_label'] = predicted_labels
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import numpy as np

# Muat kembali model dari file .h5
model = tf.keras.models.load_model("sunda_tweeter.h5")

# Muat kembali tokenizer dari file pickle
with open('tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)

# Teks yang ingin Anda prediksi
text = "aduuuhhh... hate kuring meuni bagja pisan jeung adem ngadangukeun tembang na"

# Tokenisasi teks
sequences = tokenizer.texts_to_sequences([text])

padded_sequences = pad_sequences(sequences)

# Lakukan prediksi
predictions = model.predict(padded_sequences)

# Ambil indeks kelas dengan probabilitas tertinggi
predicted_class_index = np.argmax(predictions)

# Output prediksi
print("Prediksi kelas:", predicted_class_index)
```
